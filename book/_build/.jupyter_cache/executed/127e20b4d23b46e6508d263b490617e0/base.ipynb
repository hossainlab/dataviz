```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
plt.plot(x, np.sin(x))

plt.show()
```

```python
plt.plot(x, np.sin(x), label='sine curve')
plt.plot(x, np.cos(x), label='cosine curve')

plt.legend()
plt.title('Playing with Plots')

plt.show()
```

```python
plt.plot(x, np.sin(x), label='sine curve', color='green')
plt.plot(x, np.cos(x), label='cosine curve', color='m')

plt.legend()
plt.title('Playing with Plots')

plt.show()
```

```python
random_array = np.random.randn(20)
```

```python
plt.plot(random_array, 
         color='green')
plt.show()
```

```python
plt.plot(random_array, 
         color='green',
         linestyle=':')
plt.show()
```

```python
plt.plot(random_array, 
         color='green',
         linestyle='--')
plt.show()
```

```python
plt.plot(random_array, 
         color='green',
         linestyle='--',
         linewidth=3)
plt.show()
```

```python
plt.plot(random_array,
         color='green', 
         marker = 'd')
plt.show()
```

```python
plt.plot(random_array,
         color='green', 
         marker = 'd',
         markersize=10)
plt.show()
```

```python
plt.plot(random_array,
         color='green', 
         marker = 'd',
         linestyle = 'None')
plt.show()
```

```python
plt.scatter(range(0,20),
            random_array,
            color='green', 
            marker = 'd')
plt.show()
```

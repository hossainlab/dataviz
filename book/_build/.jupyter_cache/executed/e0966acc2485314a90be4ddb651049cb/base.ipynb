```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
np.random.seed()
```

```python
y = np.random.uniform(low=0.0, high=1000, size=(1000,))
y.sort()

x = np.arange(len(y))
```

```python
plt.plot(x, y)

plt.grid(True)

plt.yscale('linear')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log', basey=2)

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log', basey=10)

plt.xscale('log', basex=2)

plt.show()
```

```python
y = 100 * np.random.normal(size=1000)
y.sort()

x = np.arange(len(y))
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('linear')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('symlog')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('symlog', linthreshy=10)

plt.show()
```

```python
y = np.random.uniform(low=0.0, high=1.0, size=(1000,))
y.sort()

x = np.arange(len(y))
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('linear')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')

plt.show()
```

```python
from matplotlib.ticker import NullFormatter
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('logit')
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.show()
```
